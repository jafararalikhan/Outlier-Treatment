#Import Dependency

```python
import pandas as pd
import math
```

#Create DataFrame

```python
df = pd.DataFrame({'column1':[50,51,56,70,100,10000]})
df
```

#Print Box Plot

```python
df['column1'].plot(kind='box')
```

#Calculate Square Root And Create a Root Column

```python
df['column_Sqr'] = df['column1'].apply(lambda x : x**(1/2))
df
```

#Calculate Cube Root And Create Clube Column

```python
df['column_cube_root'] = df['column1'].apply(lambda x: x**(1/3))
df
```

#Calculate Log Function And Create Log Column

```python
df['column_log'] = df['column1'].apply(lambda x: math.log10(x))
df
```
